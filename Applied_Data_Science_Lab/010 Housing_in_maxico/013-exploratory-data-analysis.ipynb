```python
import matplotlib.pyplot as plt
import pandas as pd
import plotly.express as px
```

```python
df=pd.read_csv('Data/mexico-real-estate-clean.csv')
df['area_m2']=df['area_m2'].astype(float)
df.info()
```

```python
fig=px.scatter_mapbox(
    df, # Our dataframe
    lat="lat",
    lon="lon",
    center={"lat":19.43,"lon":-99.13}, #map will be centered on Mexico city.
    width=600,
    height=600,
    hover_data=["price_usd"] #Display price when having mouse over house    
)

# Add mapbox_style to figure layout
fig.update_layout(mapbox_style="open-street-map")

# Show figure
fig.show()
```

```python
# Use Matplotlib to create boxplot of "area_m2"
plt.hist(df['area_m2'], rwidth=0.9)

# Add x-axis label
plt.xlabel("Area [sq meters]")

# Add y-axis label
plt.ylabel("Frequency")

# Add title
plt.title("Distribution of Home Sizes");
```

```python
# Use Matplotlib to create boxplot of "area_m2"
plt.boxplot(df['area_m2'], vert=False)

# Add x-axis label
plt.xlabel("Area [sq meters]")

# Add title
plt.ylabel("Distribution of Home Sizes");
```

```python
# Use Matplotlib to create histogram of "price_usd"
plt.hist(df['price_usd'],rwidth=0.9)

# Add x-axis label
plt.xlabel("Price [USD]")

# Add y-axis label
plt.ylabel("Frequency")

# Add title
plt.title("Distribution of Home Prices")
```

```python
# Use Matplotlib to create boxplot of "price_usd"
plt.boxplot(df["price_usd"], vert=False)

# Add x-label axis
plt.xlabel("Price [USD]")

# Add title
plt.title("Distribution of Home Prices")
```
